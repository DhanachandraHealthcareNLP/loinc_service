```python
import json
import os
import requests
from tqdm import tqdm

def get_loinc_entity_output(doc):
    url = "http://localhost:3001/loinc_output"
    with open(doc, 'r', encoding='utf8') as fopen:
        text = fopen.read()
    data = json.dumps({"content": text})
    res = requests.post(url, headers={'Content-type': 'application/json'}, data=f'{data}')
    return json.loads(res.text)
```

```python
import os 
import pandas as pd
import json

java_path = "D:/Work_Files/LOINC/testing_files/output_java_loinc/"
python_path = "D:/Work_Files/LOINC/testing_files/output_py_loinc_ner_1_server/"

files = os.listdir(java_path)

all_results = []

for f in files : 
    py_outputs = json.load(open(python_path + f))
    java_outputs = json.load(open(java_path + f))

    pyset = set()
    javaset = set()

    for out in py_outputs["result"]: 
        pyset.add(out["code"])

    for out in java_outputs["codes"]: 
        javaset.add(out["value"])

    if len(pyset) == len(javaset) == 0: 
        continue
    
    in_java_not_py = javaset.difference(pyset)
    in_py_not_java = pyset.difference(javaset)

    if len(in_java_not_py) == 0: 
        continue

    common = javaset.intersection(pyset)

    not_py_ent = []
    for val in in_java_not_py:
        for out in java_outputs["codes"]: 
            if val == out["value"] : 
                not_py_ent.append(out)


    all_results.append([f, common, in_java_not_py, in_py_not_java, not_py_ent])
```

```python
resuts_df = pd.DataFrame(all_results, columns=["File Name", "Common data", "Absent in python", "Absent in Java", "Absent python entity"])
```

```python
from typing import List
import copy
def get_combination(lisOfList):
        allCombinationList = []
        generateValues(lisOfList, "", allCombinationList)
        return allCombinationList

def generateValues(
    outerList: List[List[int]], outPut: str, allCombinationList: List[str]
):
    lst = next(iter(outerList))
    for s in lst:
        newOuter = copy.deepcopy(outerList)
        if lst in outerList:
            newOuter.remove(lst)
        if len(outerList) > 1:
            generateValues(
                newOuter, (outPut + " " + str(s)).strip(), allCombinationList
            )
        else:
            lemmaValues = (outPut + " " + str(s)).strip()

            allCombinationList.append(lemmaValues)
```

```python
import os 
import json
import pandas as pd

java_path = "D:/Work_Files/LOINC/server_upload_ner/output_ner_1_java/output_ner_1_java/"
python_path = "D:/Work_Files/LOINC/server_upload_ner/output_ner_1_corrected/output_ner_1/"
```

```python
df = pd.read_csv("Evaluation.csv")
total_words = set()
for row in df.itertuples(index=False): 
    fname = row[1]
    python_loinc_entities = eval(row[-1])

    python_entites = json.load(open(python_path + fname))


    words = []
    for ent in python_loinc_entities: 
        for env in ent["evidence"]: 
            words.append(env["textSpan"]["text"])

    py_words = []

    for ent in python_entites["result"]["entities"]: 
        for text in ent["textSpan"]: 
            py_words.append(text["text"])

    for word in words : 
        if word not in py_words: 
            words.remove(word)

    for word in words : 
        total_words.add(word)
```

```python
result = []
for word in total_words: 
    found = False
    for row in df.itertuples(index=False): 
        fname = row[1]
        python_loinc_entities = eval(row[-1])
        
        for ent in python_loinc_entities: 
            for env in ent["evidence"]: 
                if env["textSpan"]["text"] == word : 
                    result.append([fname, ent["value"], ent])
                    found = True
                    break 
                if found: 
                    break 

            if found: 
                break

        if found : 
            break  
        
```

```python
import copy

def get_labdata_values(text_dict):
    new_entites = []
    old_entities = text_dict["result"]["entities"]
    relations = text_dict["result"]["relations"]

    for ent in old_entities : 
        unit_list = []
        method_list = []
        value_list = []
        system_list = []

        ent_id = ent["id"]

        for rel in relations: 
            if rel["head"]["id"] == ent_id : 
                tail_type = rel["tail"]["type"]
                ttype = tail_type.split("_")[0]
                if ttype == "VALUE" : 
                    value_list.append(rel["tail"]["id"])
                if ttype == "METHOD" :
                    method_list.append(rel["tail"]["id"]) 
                if ttype == "SYSTEM" : 
                    system_list.append(rel["tail"]["id"])
                if ttype == "UNIT" :
                    unit_list.append(rel["tail"]["id"]) 
                
        new_ent = copy.deepcopy(ent)
        new_ent["metadata"]["labData"]["unit"] = unit_list
        new_ent["metadata"]["labData"]["value"] = value_list
        new_ent["metadata"]["labData"]["system"] = system_list
        new_ent["metadata"]["labData"]["method"] = method_list

        new_entites.append(new_ent)

    text_dict["result"]["entities"] = new_entites

    return text_dict
```
